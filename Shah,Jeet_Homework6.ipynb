```python
import numpy as np
import pandas as pd
import bqplot.pyplot
import ipywidgets as ipy
```

```python
buildings = pd.read_csv('/Users/jeet/Downloads/building_inventory.csv',na_values={'Square Footage':0,'Year Acquired':0,'Year Constructed':0})
```

### Created a buildings dataset.

```python
buildings
```

```python
build = pd.pivot_table(buildings, index = 'Agency Name', columns = 'County', values = 'Square Footage', aggfunc = 'mean', fill_value=0)

print(build)
```

## Created a pivot table from the main dataset, removed the Null values and then stored it in another variable. Took Index as 'Agency Name', Column as 'County'and Values as 'Square Footage'. Then performed aggregate function of 'mean' on the Square Footage values for each column. 

```python
#bar_plot
x_scs = bqplot.OrdinalScale()
y_scs = bqplot.LinearScale() 


x_axs = bqplot.Axis(label='Year', scale=x_scs)
y_axs = bqplot.Axis(label='log(Square Footage)', scale=y_scs, 
                   orientation = 'vertical')


barpl = bqplot.pyplot.bar(x=[], 
                         y = [],
                         scales={'x':x_scs, 'y':y_scs})

fig_barpl = bqplot.Figure(marks=[barpl], axes=[x_axs,y_axs])
```

### Set the scales for Bar Plot. For the x-axis I used Ordinal Scale as it changes all kinds of values into all kinds of values. Similarly, for y-axis I used Linear Scale as we are dealing with numerical values. 
### Next, set the Axis of Bar plot. X-axis is the Year & Y-axis is the log of Square Footage. 
### Then, created a Bar Plot keeping the X and Y values empty for now. The x & y values will be passed on later when the function is called. Lastly, thge Bar plot is plotted.

```python
#Heatmap
def barp(a,b):
    df=buildings[(buildings['County']== b) & (buildings['Agency Name']== a)]
    df= df[df['Year Acquired']!=0]
    df=df.groupby('Year Acquired')['Square Footage'].sum()
    return df


col_sc = bqplot.ColorScale(scheme="Red", 
                           min=np.nanmin(build), 
                           max=np.nanmax(build))
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()

# (2) create axis - for colors, x & y
c_ax = bqplot.ColorAxis(scale = col_sc, 
                        orientation = 'vertical', 
                        side = 'right')

x_ax = bqplot.Axis(scale = x_sc, label='County')
y_ax = bqplot.Axis(scale = y_sc, 
                   orientation = 'vertical', 
                   label = 'Agency Name')

# (3) Marks
heat_map = bqplot.GridHeatMap(color = build.values,
                              row=build.index, column = build.columns,
                              scales = {'color': col_sc,
                                        'row': y_sc,
                                        'column': x_sc},
                              interactions = {'click': 'select'},
                              anchor_style = {'fill':'blue'}, 
                              selected_style = {'opacity': 1.0},
                              unselected_style = {'opacity': 1.0})

# (4) interactivity - label
mySelectedLabel = ipy.Label()
def get_data_value(change):
    if len(change['owner'].selected) == 1: #only 1 selected
        i,j = change['owner'].selected[0]
        v = build.values[i,j] # grab data value
        mySelectedLabel.value = 'Mean Square Footage= ' + str(v) 
        g= barp(build.index[i],build.columns[j])
        barpl.x = g.index
        barpl.y = g.values
    
# make sure we check out     
heat_map.observe(get_data_value, 'selected')


# (5) put it all together in a figure
fig = bqplot.Figure(marks = [heat_map], axes = [c_ax, y_ax, x_ax])
fig.layout.min_width = '700px'
fig_barpl.layout.min_width = '700px'

myDashboard1 = ipy.HBox([fig,fig_barpl])
myDashboard = ipy.VBox([mySelectedLabel,myDashboard1])

myDashboard             
```

#### Next, I created a Gridmap with the color scheme of Heatmap. The gridmap maps the Mean Square footage of every Agency in every County. Also, when a specific block is selected, the mean Square Footage for that block. Each block represents an Agency and a County.The heatmap ranges from Red to Green. Most red blocks are 0 values. Each block is colored according to it's value on the Heatmap. Eg: 0 block values are Red, Blocks with values around 9M are colored Green and so on. 
#### For the interactivity, I used the Observe function. On every click by the user, the Observe function gets called and makes the required updates to the Visualization. Created a function to pass values to the Bar Plot. In that function, I checked for the Agency Name and County in the original dataset and got the sum for the Square Footage. Also, I deleted all the rows which had Year Acquired = 0. 
#### Lastly using the HBox function in ipywidgets library, I created the dashboard.
